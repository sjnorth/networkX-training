# Social Network Analysis: NetworkX intro 

```python
import networkx as nx
```

```python
G = nx.Graph() #this variable stores all the nodes and edges for our graph. 
```

```python
#these are inbuilt methods within Graph
G.add_node('bob') 
G.add_node('carole')
G.add_node('steve')
```

```python
G.add_edge('bob', 'carole')
G.add_edge('carole', 'steve') #add relationship between these two nodes 
```

```python
nx.density(G) #Observed edges / total possible edges 
```

```python
#you might need to run this twice if you don't see anything displayed
import matplotlib.pyplot as plt
nx.draw_networkx(G) #networkx has various functions which operate on graphs
plt.show()
```

```python
#loop through nodes 
for node in G.nodes:
    print(node)
```

```python
G.add_node('bob', color='blue')
```

```python
G.add_edge('bob', 'carole', weight=1)
```
